```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
target = 21

pool = [11,2,3,4,5,6,7,8,9,10,10,10,10]
counts = []
on_targets = []
for i in range(10000):
    total_sum = 0
    count = 1
    while total_sum < target:
        total_sum += np.random.choice(pool)
        count += 1
    on_target = True if total_sum == target else False
    counts.append(count)
    on_targets.append(on_target)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=counts,hue=on_targets)
plt.xlabel(f'Number of draws needed before {target} is reached')
plt.tight_layout()
plt.show()
```
